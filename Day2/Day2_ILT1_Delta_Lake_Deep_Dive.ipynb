## Section 1 — ACID Transactions — What Each Letter Actually Means

You heard the word "ACID" yesterday. Today we understand what it means in real life.

---

### A — Atomicity — "All or Nothing"

When you write data, either **ALL rows are saved** or **NONE of them are**.

**Real example:**
```
You are writing 10,000 order rows to Bronze.
Your cluster crashes after writing row 7,432.

Without ACID (CSV):  You have 7,432 rows. Data is CORRUPTED.
With ACID (Delta):   Delta sees the write was incomplete → rolls it back.
                     You have 0 rows. Data is CLEAN. Run again.
```

---

### C — Consistency — "Rules Are Always Followed"

The table is always in a valid state — no half-written data, no schema violations.

**Real example:**
```
Your orders table has a schema: order_id (string), order_status (string)
You try to write a row where order_id = 12345 (integer, not string)

Delta refuses the write and raises an error.
The table stays consistent — no mixed types.
```

---

### I — Isolation — "Readers and Writers Don't Block Each Other"

Two notebooks can read and write at the same time — they don't interfere.

**Real example:**
```
Notebook A:  Writing new orders to Bronze (takes 5 minutes)
Notebook B:  Analyst reading orders at the SAME TIME

Without ACID (CSV):  Analyst reads half-written data. Garbage results.
With ACID (Delta):   Analyst reads the PREVIOUS version while you write.
                     As soon as your write finishes, they see the new version.
```

---

### D — Durability — "Once Written, Always There"

Once Delta confirms the write is done, the data will not disappear — even if the cluster crashes.

**Real example:**
```
You write 10,000 rows. Databricks shows "Write complete".
Your cluster crashes 1 second later.

Without durability:  Data might be in memory, not yet flushed to disk.
With ACID (Delta):   Data is written to ADLS Gen2 cloud storage.
                     It is there permanently. Cloud storage = durable.
```

## Section 2 — Inside the `_delta_log` Folder

Every Delta table has a hidden folder called `_delta_log`.  
This is where Delta keeps the **transaction history** — every single write is recorded here.

### What is inside `_delta_log`?

```
bronze/
├── orders/
│   ├── _delta_log/
│   │   ├── 00000000000000000000.json   ← first write (version 0)
│   │   ├── 00000000000000000001.json   ← second write (version 1)
│   │   ├── 00000000000000000002.json   ← third write (version 2)
│   │   └── 00000000000000000010.checkpoint.parquet  ← checkpoint every 10 versions
│   ├── part-00000-abc123.snappy.parquet  ← actual data files
│   └── part-00001-def456.snappy.parquet
```

### What is inside one of those JSON files?

Here is a simplified version of what `00000000000000000001.json` looks like:

```json
{
  "add": {
    "path": "part-00000-abc123.snappy.parquet",
    "size": 245000,
    "modificationTime": 1706783400000,
    "stats": "{\"numRecords\": 3000, \"minValues\": {\"order_purchase_timestamp\": \"2018-01-01\"}}"
  }
}
```

Every JSON file has one or more of these operations:

| Operation | Meaning |
|-----------|--------|
| `add` | A new data file was written |
| `remove` | An old data file was logically deleted (still on disk until VACUUM) |
| `metaData` | Schema information was saved |
| `commitInfo` | Who wrote, when, what operation (INSERT, UPDATE, DELETE) |

> **Key insight:** Delta does NOT delete files immediately.  
> It just marks old files as `remove` in the log.  
> This is how time travel works — old files are still there!

> **Live demo tip:** Go to Databricks → Data → navigate to your Delta table → browse the _delta_log folder to show students the actual JSON files.

## Section 3 — Viewing the Transaction Log

Let us look inside the `_delta_log` folder to see what Delta actually writes.

````python
# ─── Look inside the _delta_log folder ────────────────────────────────────────
# This shows all the JSON files that Delta created when we wrote orders

delta_log_path = f"{bronze_path}/orders/_delta_log"

print("Files inside _delta_log for the orders table:")
print("-" * 60)

try:
    log_files = dbutils.fs.ls(delta_log_path)
    for f in log_files:
        print(f"  {f.name}")
    print(f"\nTotal log entries: {len(log_files)}")
except Exception as e:
    print(f"Note: {e}")
    print("Make sure you have written orders to Bronze first (from yesterday's hands-on)")
````

````python
# ─── Read the contents of a transaction log JSON file ─────────────────────────
# This shows what Delta wrote when we saved the orders table

import json

# Read the very first log entry (version 0 = the first write)
first_log_file = f"{delta_log_path}/00000000000000000000.json"

try:
    log_content = dbutils.fs.head(first_log_file, 2000)  # read first 2000 characters
    
    print("Contents of version 0 log (first write):")
    print("-" * 60)
    
    # Each line in the JSON file is one operation
    for line in log_content.strip().split("\n"):
        if line.strip():
            entry = json.loads(line)
            operation_type = list(entry.keys())[0]  # add, remove, commitInfo, metaData
            print(f"\nOperation: '{operation_type}'")
            
            if operation_type == "commitInfo":
                info = entry["commitInfo"]
                print(f"  Timestamp  : {info.get('timestamp', 'N/A')}")
                print(f"  Operation  : {info.get('operation', 'N/A')}")
            elif operation_type == "add":
                add_info = entry["add"]
                print(f"  File added : {add_info.get('path', 'N/A')}")
                print(f"  File size  : {add_info.get('size', 0) / 1024:.1f} KB")

except Exception as e:
    print(f"Could not read log file: {e}")
````

````python
# ─── Check Delta Table History ────────────────────────────────────────────────
# DESCRIBE HISTORY is the easy way to see all versions of a Delta table

# First, create a temporary view so we can use SQL
orders_delta = spark.read.format("delta").load(f"{bronze_path}/orders")
orders_delta.createOrReplaceTempView("bronze_orders")

# DESCRIBE HISTORY shows every version: when it was written, what operation, by whom
print("Full version history of the orders Delta table:")
spark.sql("DESCRIBE HISTORY bronze_orders").select(
    "version", "timestamp", "operation", "operationParameters"
).show(truncate=False)
````

## Section 4 — Time Travel — Query Data from the Past

Because Delta keeps old files (it only marks them `remove`, never deletes them),  
you can query the table **as it was at any past version or timestamp**.

### Why is this useful?

| Use Case | How Time Travel Helps |
|----------|-----------------------|
| Someone ran a bad pipeline and corrupted data | Restore from the previous version |
| Auditor wants to know what the data looked like last month | Query by timestamp |
| You want to compare how a table changed over time | Compare version 0 vs version 5 |
| You accidentally dropped rows | Roll back to before the drop |

### Two ways to time travel

```python
# Option 1 — By version number
df = spark.read.format("delta").option("versionAsOf", 0).load(path)

# Option 2 — By timestamp
df = spark.read.format("delta").option("timestampAsOf", "2024-01-15").load(path)
```

````python
# ─── Write new data to create multiple versions ────────────────────────────────
# We will write to the orders table 3 times to create 3 versions
# This shows how time travel works

# Read orders from raw CSV
orders_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{raw_path}/orders.csv")

print(f"Total orders in source CSV: {orders_df.count()}")
orders_df.printSchema()
````

````python
# ─── Write Version 0 — First write ─────────────────────────────────────────────
# Write only orders from 2024 — this is Version 0
# Adjust year value if your dataset uses different years

from pyspark.sql.functions import col, to_timestamp, year

orders_with_year = orders_df.withColumn(
    "OrderDate",
    to_timestamp(col("OrderDate"))
).withColumn("order_year", year(col("OrderDate")))

# Filter: only 2024 orders
orders_2024 = orders_with_year.filter(col("order_year") == 2024)

print(f"Writing Version 0: orders from 2024 ({orders_2024.count()} rows)")

orders_2024.write \
    .format("delta") \
    .mode("overwrite") \
    .save(f"{bronze_path}/orders_demo")

print("Version 0 written!")
````

````python
# ─── Write Version 1 — Add 2025 orders ─────────────────────────────────────────
# Append 2025 orders — this creates Version 1

orders_2025 = orders_with_year.filter(col("order_year") == 2025)

print(f"Writing Version 1: appending 2025 orders ({orders_2025.count()} rows)")

orders_2025.write \
    .format("delta") \
    .mode("append") \
    .save(f"{bronze_path}/orders_demo")

print("Version 1 written!")
print(f"\nTotal rows now: {spark.read.format('delta').load(f'{bronze_path}/orders_demo').count()}")
````

````python
# ─── TIME TRAVEL — Read Version 0 (only 2024 data) ──────────────────────────
# Even though we appended 2025 data, we can still go back to the original state!

demo_path = f"{bronze_path}/orders_demo"

# Read Version 0 — only 2024 data
v0 = spark.read \
    .format("delta") \
    .option("versionAsOf", 0) \
    .load(demo_path)

# Read Version 1 — 2024 + 2025 data
v1 = spark.read \
    .format("delta") \
    .option("versionAsOf", 1) \
    .load(demo_path)

# Compare the two versions
print("Time Travel Comparison:")
print(f"  Version 0 (2024 only):      {v0.count():,} rows")
print(f"  Version 1 (2024 + 2025):    {v1.count():,} rows")
print(f"  Rows added in Version 1:    {v1.count() - v0.count():,} rows")

print("\nVersion 0 — sample rows:")
v0.select("OrderID", "OrderChannel", "OrderDate").show(3)

print("Version 1 — sample rows (includes 2025):")
v1.select("OrderID", "OrderChannel", "OrderDate").show(3)
````

````python
# ─── TIME TRAVEL — Restore a table to a previous version ──────────────────────
# Scenario: someone accidentally overwrote good data with bad data
# We want to go back to Version 0

# Step 1: Read the good version
good_version = spark.read \
    .format("delta") \
    .option("versionAsOf", 0) \
    .load(demo_path)

print(f"Restoring to Version 0 ({good_version.count()} rows)...")

# Step 2: Overwrite the current table with the old version
good_version.write \
    .format("delta") \
    .mode("overwrite") \
    .save(demo_path)

restored = spark.read.format("delta").load(demo_path)
print(f"Table restored! Current row count: {restored.count()}")

# This creates Version 2 — a restore operation
# The history now shows: V0 (insert) → V1 (append) → V2 (restore)
print("\nFull history after restore:")
restored.createOrReplaceTempView("orders_demo")
spark.sql("DESCRIBE HISTORY orders_demo").select(
    "version", "timestamp", "operation"
).show()
````

## Section 5 — VACUUM — Cleaning Up Old Files

Remember: Delta does NOT immediately delete old files.  
It just marks them as `remove` in the transaction log.  
Over time, this means **many old Parquet files pile up on disk**.

### The Problem Without VACUUM

```
After 100 writes to a Delta table, you might have:
  - 5 current data files  (what the table looks like NOW)
  - 95 old data files     (needed only for time travel)

All 100 files are sitting in ADLS, costing you storage money!
```

### VACUUM — What it does

VACUUM deletes old data files that are **no longer needed** for time travel.

```python
VACUUM delta_table RETAIN 168 HOURS   # keep 7 days of history (default)
VACUUM delta_table RETAIN 0 HOURS     # delete everything older than now (DANGEROUS!)
```

### Important Rules for VACUUM

| Rule | Why |
|------|-----|
| Default retention is 7 days (168 hours) | Safe for most use cases |
| Cannot time travel past the VACUUM point | Once files are deleted, that version is gone |
| Do NOT run VACUUM RETAIN 0 HOURS on production | You will lose ALL history |
| Run VACUUM regularly (weekly is good) | Keeps storage costs low |

````python
# ─── VACUUM — Clean up old files ──────────────────────────────────────────────
# This removes old Parquet files that are no longer needed

# First, let us see how many files are in the orders_demo folder BEFORE vacuum
all_files_before = dbutils.fs.ls(demo_path)

# Count only data files (Parquet files, not the _delta_log folder)
parquet_files_before = [f for f in all_files_before if f.name.endswith(".parquet")]
print(f"Parquet files BEFORE VACUUM: {len(parquet_files_before)}")

# Run VACUUM — keep 0 hours for demo purposes
# WARNING: In production, always use at least 168 hours (7 days)
# We use 0 here just to show the effect in the classroom
spark.conf.set("spark.databricks.delta.retentionDurationCheck.enabled", "false")

spark.sql(f"VACUUM delta.`{demo_path}` RETAIN 0 HOURS")

# Count files AFTER vacuum
all_files_after = dbutils.fs.ls(demo_path)
parquet_files_after = [f for f in all_files_after if f.name.endswith(".parquet")]
print(f"Parquet files AFTER VACUUM:  {len(parquet_files_after)}")
print(f"Files deleted: {len(parquet_files_before) - len(parquet_files_after)}")

# Re-enable the safety check for future use
spark.conf.set("spark.databricks.delta.retentionDurationCheck.enabled", "true")
````

## Section 6 — OPTIMIZE — Fixing the Small Files Problem

### What is the "Small Files Problem"?

When you write data many times (especially with `mode("append")`), Spark creates many small files.  
Reading many small files is **slow** — Spark has to open each file one by one.

```
After 50 append writes to orders table:

  part-00001.parquet  (5 KB)
  part-00002.parquet  (3 KB)
  part-00003.parquet  (7 KB)
  ... (50 files, each tiny)

Reading this = Spark opens 50 files × overhead per file = SLOW
```

### OPTIMIZE — What it does

OPTIMIZE **merges small files** into a few large files (target: ~128 MB each).  
This makes reads much faster.

```
Before OPTIMIZE:  50 files × 5 KB = 250 KB total
After OPTIMIZE:   1 file × 250 KB  = 250 KB total  (same data, fewer files!)
```

````python
# ─── OPTIMIZE — Compact small files ───────────────────────────────────────────

# First, let us see how many files are in the orders Bronze table BEFORE optimize
orders_path = f"{bronze_path}/orders"

try:
    files_before = dbutils.fs.ls(orders_path)
    parquet_before = [f for f in files_before if f.name.endswith(".parquet")]
    print(f"Files BEFORE OPTIMIZE: {len(parquet_before)}")
    for f in parquet_before[:5]:  # show first 5 files
        print(f"  {f.name} ({f.size / 1024:.1f} KB)")
except Exception as e:
    print(f"Note: {e} — using orders_demo path instead")
    orders_path = demo_path

# Run OPTIMIZE on the table
optimize_result = spark.sql(f"OPTIMIZE delta.`{orders_path}`")
print("\nOPTIMIZE complete!")
optimize_result.show()

# Files AFTER optimize
files_after = dbutils.fs.ls(orders_path)
parquet_after = [f for f in files_after if f.name.endswith(".parquet")]
print(f"\nFiles AFTER OPTIMIZE: {len(parquet_after)}")
for f in parquet_after:
    print(f"  {f.name} ({f.size / 1024:.1f} KB)")
````

## Section 7 — Z-order — Making Queries Faster

### What is Z-ordering?

Z-ordering is a way to **sort the data inside Parquet files** so that queries on certain columns are much faster.

### How does normal reading work?

```
Query: SELECT * FROM orders WHERE customer_id = 'C001'

Without Z-order:
  Spark reads file 1 → scan all rows → some match
  Spark reads file 2 → scan all rows → some match
  Spark reads file 3 → scan all rows → none match (wasted effort!)
  Spark reads file 4 → scan all rows → some match
  ... reads ALL files
```

### How Z-order helps

```
After Z-order on customer_id:
  File 1: customer_id A001 – B999 (min/max stored in metadata)
  File 2: customer_id C000 – D999
  File 3: customer_id E000 – Z999

Query: WHERE customer_id = 'C001'
  Spark checks metadata → only File 2 can have C001
  Reads ONLY File 2 → skips Files 1 and 3 completely!
```

This is called **data skipping** — skipping files that cannot possibly match your filter.

### When to use Z-order?

| Scenario | Z-order column |
|----------|---------------|
| Frequently filter by customer | `customer_id` |
| Frequently filter by date range | `order_purchase_timestamp` |
| Frequently filter by seller | `seller_id` |
| Frequently filter by product | `product_id` |

> **Rule:** Z-order columns should be columns you filter on most often in your WHERE clause.

````python
# ─── OPTIMIZE with Z-order ────────────────────────────────────────────────────
# Z-order the orders table by CustomerID
# This makes queries like "WHERE CustomerID = 'X'" much faster

print("Running OPTIMIZE with Z-order on CustomerID...")
print("This sorts data inside Parquet files so similar CustomerIDs are close together.")
print()

# Run OPTIMIZE with ZORDER
# Only one Z-order column recommended — more columns = diminishing returns
result = spark.sql(f"""
    OPTIMIZE delta.`{orders_path}`
    ZORDER BY (CustomerID)
""")

print("OPTIMIZE + ZORDER complete!")
result.show()
print("\nNow queries like 'WHERE CustomerID = ...' will be much faster!")
````

````python
# ─── See the effect of Z-order using EXPLAIN ──────────────────────────────────
# EXPLAIN shows Spark's query plan — look for "PartitionFilters" and "numFiles"

orders_after_zorder = spark.read.format("delta").load(orders_path)
orders_after_zorder.createOrReplaceTempView("orders_zordered")

# Query with a CustomerID filter — after Z-order, Spark skips irrelevant files
# The query itself looks the same — the speed difference is internal
result = spark.sql("""
    SELECT OrderID, CustomerID, OrderChannel, OrderDate
    FROM orders_zordered
    WHERE CustomerID = 'CUST-08966'
""")

result.show(truncate=False)
print(f"Rows found: {result.count()}")
print("\nWith Z-order, Spark reads fewer files to find this customer — faster!")
````

## Summary — OPTIMIZE vs VACUUM vs Z-order

| Command | What it does | When to run |
|---------|-------------|-------------|
| `OPTIMIZE` | Merge small files into big files | After many small writes (daily/weekly) |
| `VACUUM` | Delete old files no longer needed | Weekly, retain at least 7 days |
| `OPTIMIZE ZORDER BY` | Sort data inside files for fast filtering | When you add OPTIMIZE, pick your key filter column |

### The Recommended Maintenance Pattern

```python
# Run this weekly on each Delta table:

# Step 1: Compact files + sort for fast reads
OPTIMIZE delta.`path/to/table` ZORDER BY (most_filtered_column)

# Step 2: Delete old files (keep 7 days)
VACUUM delta.`path/to/table` RETAIN 168 HOURS
```
