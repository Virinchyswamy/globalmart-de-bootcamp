## Section 1: What is Idempotency?

**INSTRUCTOR NOTE:**  
Use the ATM analogy first. Students will get it immediately without needing a technical definition.

---

### The ATM Analogy:

Imagine you go to an ATM and withdraw ₹1000.  
The ATM shows an error — you are not sure if the transaction went through.  
You press the button again.

**Bad ATM (not idempotent):** Deducts ₹1000 twice. You lose ₹2000 even though you only wanted ₹1000.  
**Good ATM (idempotent):** Checks if the first transaction already happened. If yes, it does not deduct again. You get ₹1000 exactly once — no matter how many times you pressed the button.

---

### Definition:

> **Idempotent** = Running the same operation multiple times produces the same result as running it once.

In data pipelines:
- Your pipeline runs every day at 6 AM
- One day it fails halfway and gets restarted
- **Non-idempotent pipeline:** the data gets duplicated or partially loaded
- **Idempotent pipeline:** rerunning it gives exactly the same correct result — no duplicates, no partial loads

---

### Why Pipelines Fail and Get Rerun:

| Reason | Example |
|--------|--------|
| **Cluster crash** | Databricks cluster runs out of memory and dies mid-job |
| **Network timeout** | ADLS connection drops while writing |
| **Source data delay** | Supabase Postgres was slow, job timed out |
| **Manual rerun** | Engineer reruns yesterday's pipeline to fix bad data |
| **Scheduled retry** | Orchestrator automatically retries failed jobs |

**In all these cases: your pipeline MUST be safe to run again.**

**INSTRUCTOR NOTE:**  
Ask: *'Imagine your pipeline runs at 6 AM and loads 10,000 orders. It crashes at 9,000 and restarts. If it is not idempotent, what happens?'*  
Answer: You get 9,000 duplicate rows from the first run + 10,000 rows from the second run = 19,000 rows. Reports are wrong.

## Section 2: Non-Idempotent vs Idempotent — Side by Side

**INSTRUCTOR NOTE:**  
Show the code comparison below. This is the most important part of the session.

---

### Example: Loading orders into Silver

**NON-IDEMPOTENT (DANGEROUS — DO NOT USE):**
```python
# mode('append') adds rows every time it runs
# Run it 3 times = 3 copies of the same data in Silver
df_orders.write \
    .format('delta') \
    .mode('append') \
    .save(silver_orders_path)
```

**What goes wrong:** If this runs twice, you get 2x rows. If it runs 10 times, you get 10x rows.  
Your revenue reports will show 10x the real revenue. Business will panic.

---

**IDEMPOTENT (SAFE — USE THIS):**
```python
# mode('overwrite') replaces the entire table every time
# Run it 3 times = same correct result every time
df_orders.write \
    .format('delta') \
    .mode('overwrite') \
    .save(silver_orders_path)
```

**Why it is safe:** No matter how many times you run it, the result is always the latest correct data.

---

### When to use each mode:

| Mode | What it does | Use when |
|------|-------------|----------|
| `overwrite` | Replaces ALL existing data | Full reload — replace everything with latest data |
| `append` | Adds new rows on top of existing | Only when you are 100% sure no duplicates can occur |
| `merge` (UPSERT) | Update existing rows, insert new ones | Incremental loads — add new + update changed rows |

**For beginners: always use `overwrite` until you learn MERGE.**

## Section 3: The 3 Idempotent Patterns

**INSTRUCTOR NOTE:**  
Explain each pattern briefly. The code demos below show all 3 with GlobalMart data.

---

### Pattern 1: Full Overwrite
- Reload the entire table from scratch every time
- Simple, safe, always correct
- Works well for small-to-medium tables
- Not efficient for very large tables (rewrites everything even if only 10 rows changed)

### Pattern 2: Deduplicate Before Write
- Remove duplicate rows before writing
- Useful when the source data itself has duplicates
- Use `dropDuplicates(['unique_column'])` to keep only one row per unique ID

### Pattern 3: MERGE (Upsert)
- The most powerful pattern
- Update rows that already exist, insert rows that are new
- No duplicates possible because MERGE checks the key first
- Used for incremental loads (only process new/changed rows)
- We cover this deeply in Week 2 (SCD Type 1 and Type 2)

````python
# ============================================================
# CELL 1: Setup — connect to ADLS
#
# INSTRUCTOR NOTE:
#   Standard setup. Paste your key and run.
# ============================================================

storage_account_name = "amazonprojectadls"
container_name       = "amazon-data"
storage_account_key  = "YOUR_STORAGE_ACCOUNT_KEY"   # paste your key here

spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

BASE_PATH   = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net"
BRONZE_PATH = f"{BASE_PATH}/raw"
SILVER_PATH = f"{BASE_PATH}/silver"

print("Connected to ADLS.")
print(f"Bronze: {BRONZE_PATH}")
print(f"Silver: {SILVER_PATH}")
````

````python
# ============================================================
# CELL 2: Read orders from Bronze
#
# INSTRUCTOR NOTE:
#   We read orders from the raw CSV (Bronze layer).
#   First, print the schema to see the actual column names.
#   This is always the first thing to do with any new dataset.
# ============================================================

# Read the raw orders CSV from bronze
df_orders_raw = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(f"{BRONZE_PATH}/orders.csv")
)

# Always check the schema first — tells you column names and data types
print("=== ORDERS TABLE SCHEMA ===")
df_orders_raw.printSchema()

print(f"Total rows: {df_orders_raw.count():,}")
print()
print("Sample rows:")
df_orders_raw.show(5, truncate=True)
````

````python
# ============================================================
# CELL 3: DEMO — Non-Idempotent write (append mode)
#
# INSTRUCTOR NOTE:
#   Run this cell TWICE and watch the row count double.
#   This shows exactly why 'append' is dangerous for full loads.
#
#   After running twice, the silver table has 2x the real rows.
#   Any SUM or COUNT query on this table will give WRONG results.
# ============================================================

silver_orders_path = f"{SILVER_PATH}/orders_demo_bad"

# DANGEROUS: append mode - adds rows every single time it runs
df_orders_raw.write \
    .format("delta") \
    .mode("append") \
    .save(silver_orders_path)

# Read back and count rows
df_check = spark.read.format("delta").load(silver_orders_path)
row_count = df_check.count()

original_count = df_orders_raw.count()

print(f"Original orders in source : {original_count:,}")
print(f"Rows in silver table now  : {row_count:,}")
print()

if row_count > original_count:
    print(f"PROBLEM: Table has {row_count - original_count:,} EXTRA rows from duplicate runs!")
    print("Revenue reports will be WRONG by this amount.")
else:
    print("First run - run this cell again to see the duplication problem.")
````

````python
# ============================================================
# CELL 4: Pattern 1 — Idempotent write using overwrite mode
#
# INSTRUCTOR NOTE:
#   Run this cell as many times as you want.
#   The row count will ALWAYS equal the source row count.
#   This is what 'idempotent' means - same result every time.
# ============================================================

silver_orders_safe_path = f"{SILVER_PATH}/orders_demo_good"

# SAFE: overwrite mode - replaces the entire table every time
df_orders_raw.write \
    .format("delta") \
    .mode("overwrite") \
    .save(silver_orders_safe_path)

# Read back and verify
df_check_safe = spark.read.format("delta").load(silver_orders_safe_path)
row_count_safe = df_check_safe.count()

original_count = df_orders_raw.count()

print(f"Original orders in source : {original_count:,}")
print(f"Rows in silver table now  : {row_count_safe:,}")
print()

if row_count_safe == original_count:
    print("SAFE: Row count matches exactly. Run this cell 10 more times - it will still match.")
    print("This pipeline is idempotent.")
else:
    print("Something went wrong - check the write.")
````

````python
# ============================================================
# CELL 5: Pattern 2 — Remove duplicates before writing
#
# INSTRUCTOR NOTE:
#   Sometimes the source data itself has duplicate rows.
#   For example: the same customer_id appears twice because of
#   a bug in the Supabase export.
#
#   dropDuplicates(['column_name']) keeps only the FIRST row
#   for each unique value of that column.
#
#   We deduplicate on 'order_id' because each order should
#   appear exactly once in the orders table.
#
#   IMPORTANT: Replace 'order_id' with the actual ID column
#   name from your orders.csv (check the schema output above).
# ============================================================

from pyspark.sql.functions import col

# Step 1: Check for duplicates in the raw data
total_rows    = df_orders_raw.count()
distinct_rows = df_orders_raw.dropDuplicates().count()

print(f"Total rows in orders.csv         : {total_rows:,}")
print(f"Distinct rows (no duplicates)    : {distinct_rows:,}")
print(f"Duplicate rows found             : {total_rows - distinct_rows:,}")
print()

# Step 2: Remove duplicates based on the unique order ID column
# NOTE: Check your actual column name from the schema above
#       Common names: 'order_id', 'id', 'ORDER_ID'
df_orders_deduped = df_orders_raw.dropDuplicates()   # removes exact duplicate rows

print(f"Rows after deduplication: {df_orders_deduped.count():,}")

# Step 3: Write the deduplicated data to Silver
df_orders_deduped.write \
    .format("delta") \
    .mode("overwrite") \
    .save(f"{SILVER_PATH}/orders_deduped")

print("Deduplicated orders written to silver.")
````

````python
# ============================================================
# CELL 6: Pattern 3 — MERGE (Upsert) — Preview
#
# INSTRUCTOR NOTE:
#   MERGE is the most powerful idempotency pattern.
#   It checks: 'Does this row already exist in Silver?'
#   YES -> Update the existing row with new values
#   NO  -> Insert as a new row
#
#   This means no duplicates are EVER possible.
#   Each unique order_id will only ever appear once in Silver.
#
#   We cover MERGE deeply in Week 2 (SCD Type 1 and Type 2).
#   This cell is just a preview to show the concept.
# ============================================================

from delta.tables import DeltaTable

# Path to the Silver orders table
silver_orders_path = f"{SILVER_PATH}/orders_demo_good"

# Load Delta table
silver_delta_table = DeltaTable.forPath(spark, silver_orders_path)

# Simulate incoming data
df_new_orders = df_orders_raw.limit(100)

# MERGE (UPSERT)
silver_delta_table.alias("existing") \
    .merge(
        df_new_orders.alias("incoming"),
        "existing.OrderID = incoming.OrderID"
    ) \
    .whenMatchedUpdateAll() \
    .whenNotMatchedInsertAll() \
    .execute()

print("MERGE complete.")
print("Existing rows were updated, new rows were inserted.")
print("No duplicates possible with MERGE.")
print()

print(
    f"Silver table row count after MERGE: "
    f"{spark.read.format('delta').load(silver_orders_path).count():,}"
)
````

````python
# ============================================================
# CELL 7: Final — Idempotent Bronze to Silver pipeline for ALL tables
#
# INSTRUCTOR NOTE:
#   This is what a production-ready, idempotent ingestion pipeline
#   looks like for all 10 GlobalMart tables.
#
#   Key features:
#   1. Uses overwrite mode -> safe to rerun anytime
#   2. Removes duplicate rows -> clean data in Silver
#   3. Loops through all 10 tables -> one pipeline for everything
#   4. Handles errors gracefully -> one failure does not stop the rest
# ============================================================

# All 10 GlobalMart source tables
globalmart_tables = [
    "customers",
    "orders",
    "orders_items",
    "products",
    "payments",
    "addresses",
    "returns",
    "suppliers",
    "payment_methods",
    "shipping_tier"
]

print("Running idempotent Bronze -> Silver pipeline for all tables...")
print("=" * 60)
print(f"{'Table':<25} {'Rows':>10}  Status")
print("-" * 55)

for table_name in globalmart_tables:
    try:
        # --- STEP 1: Read from Bronze (raw CSV) ---
        df = (
            spark.read
            .option("header", "true")
            .option("inferSchema", "true")
            .csv(f"{BRONZE_PATH}/{table_name}.csv")
        )

        # --- STEP 2: Remove any duplicate rows ---
        df_clean = df.dropDuplicates()

        row_count = df_clean.count()

        # --- STEP 3: Write to Silver as Delta (overwrite = idempotent) ---
        df_clean.write \
            .format("delta") \
            .mode("overwrite") \
            .save(f"{SILVER_PATH}/{table_name}")

        print(f"{table_name:<25} {row_count:>10,}  OK - written to silver")

    except Exception as e:
        # If one table fails, log the error and continue with the next table
        print(f"{table_name:<25} {'':>10}  FAILED: {str(e)[:60]}")

print("-" * 55)
print()
print("Pipeline complete. Run this cell again - you will get the same result.")
print("That is idempotency.")
````
