## Section 1: Medallion Architecture

**INSTRUCTOR NOTE:**  
Start with this question: *'We have 10 raw CSV files in ADLS. Can we just query them directly for business reports?'*  
Let students think. Answer: Technically yes, but it's messy - raw data has nulls, wrong formats, duplicates, no joins. That's why we clean it in layers.

---

**Medallion Architecture** = a design pattern where data passes through 3 quality layers before it is used by the business.

Each layer has a specific purpose and set of rules. The name comes from the medals: Bronze (raw) → Silver (clean) → Gold (refined).

---

### The Three Layers:

```
SOURCE DATA
    |
    v
+-------------------+
|   BRONZE LAYER    |  <-- Raw, as-is, never modified
|   (Landing Zone)  |
+-------------------+
    |
    v  (clean, standardize, validate)
+-------------------+
|   SILVER LAYER    |  <-- Cleaned, structured, Delta format
|   (Refined Zone)  |
+-------------------+
    |
    v  (aggregate, join, apply business rules)
+-------------------+
|    GOLD LAYER     |  <-- Business-ready, reporting-ready
|   (Business Zone) |
+-------------------+
    |
    v
BI TOOLS / REPORTS / DASHBOARDS
```

---

### Bronze Layer - The Rules:

| Rule | Explanation |
|------|-------------|
| **Store raw data as-is** | Never change the original data. If something goes wrong, you can always go back. |
| **No transformations** | Do not rename columns, fix nulls, or change formats at this layer |
| **Append only** | New data is added, old data is never deleted |
| **Keep original format** | CSV stays CSV, JSON stays JSON |
| **Add metadata columns** | Add `ingestion_timestamp`, `source_file_name` to track when/where data came from |

**Think of Bronze as:** A photocopy of the original. You always have the original to refer back to.

---

### Silver Layer - The Rules:

| Rule | Explanation |
|------|-------------|
| **Delta format only** | All silver tables are Delta tables (not CSV) |
| **Clean and standardize** | Fix nulls, correct data types, standardize formats (e.g., dates as YYYY-MM-DD) |
| **Remove duplicates** | One record per entity |
| **Apply business rules** | Validate email formats, check valid status codes, etc. |
| **Join related tables** | e.g., orders joined with customers |
| **No aggregation** | Still row-level data, not summarized |

**Think of Silver as:** Clean, trustworthy data. A data analyst can query this directly.

---

### Gold Layer - The Rules:

| Rule | Explanation |
|------|-------------|
| **Delta format only** | All gold tables are Delta tables |
| **Aggregated and summarized** | Monthly sales totals, customer segments, product rankings |
| **Business-domain language** | Column names make sense to non-technical users |
| **Optimized for reporting** | Fast to query, pre-joined, pre-grouped |
| **Curated by business need** | Each Gold table answers a specific business question |

**Think of Gold as:** The final answer. A manager can query this without knowing SQL deeply.

---

### Layer Comparison:

| | Bronze | Silver | Gold |
|-|--------|--------|------|
| **Format** | Original (CSV, JSON) | Delta | Delta |
| **Transformations** | None | Yes (clean, validate) | Yes (aggregate, join) |
| **Data Quality** | Raw (may have errors) | High | Highest |
| **Who uses it** | Data Engineers only | Engineers + Analysts | Business + BI Tools |
| **Granularity** | Raw rows | Cleaned rows | Aggregated rows |
| **Typical path** | `raw/customers.csv` | `silver/customers/` | `gold/monthly_sales/` |

## Section 2: What is Delta Lake?

**INSTRUCTOR NOTE:**  
This is the most important concept of the entire bootcamp. Spend 15 minutes here. Use the bank analogy.

---

**Delta Lake** is an open-source storage layer that adds reliability to data stored in ADLS (or any cloud storage).  
It is NOT a separate database - it works ON TOP of Parquet files already stored in ADLS, and adds a transaction log alongside them.

### The Problem Delta Lake Solves:

Without Delta Lake, data stored as CSV/Parquet files in ADLS has these problems:

| Problem | Example | Impact |
|---------|---------|--------|
| **No ACID transactions** | Two jobs write at the same time → data gets corrupted | Reports show wrong numbers |
| **No schema enforcement** | Someone uploads a CSV with extra/missing columns → job crashes | Pipeline failure |
| **No updates or deletes** | Need to fix a wrong customer address? You cannot do it on raw files easily | Stale/incorrect data |
| **No audit trail** | Who changed what and when? No way to know | No traceability |
| **Partial writes** | Job fails halfway → half the data is written, half is missing | Incomplete data |

Delta Lake solves ALL of these.

---

### The Bank Account Analogy:

Imagine a bank with NO transaction history:
- Someone credits your account - but the system crashes halfway. Did the money arrive?
- Two people withdraw at the same time - does the balance go negative?
- An error was made last week - can you go back and see the old balance?

**Without Delta Lake** = a bank with no transaction log. Chaos.

**With Delta Lake** = every write, update, delete is logged. You can roll back. You can see history. Nothing is lost.

---

### How Delta Lake Works - The Transaction Log:

When you save data as a Delta table, Delta creates a special folder called `_delta_log` alongside the data files:

```
silver/
    customers/
        _delta_log/               <-- Delta's brain
            00000000000000000000.json   (version 0 - first write)
            00000000000000000001.json   (version 1 - second write)
            00000000000000000002.json   (version 2 - update)
        part-00000-abc123.parquet   <-- actual data (stored as Parquet)
        part-00001-def456.parquet
```

Every time something changes (write, update, delete), Delta writes a new JSON entry in `_delta_log`.  
This log is what gives Delta its superpowers:

| Superpower | What it means | How it works |
|-----------|--------------|-------------|
| **ACID Transactions** | All-or-nothing writes | Transaction log is updated only on success |
| **Schema Enforcement** | Rejects bad data automatically | Schema is stored in the log |
| **Time Travel** | Read data as it was yesterday/last week | Each version is stored in the log |
| **Audit History** | See every change ever made | Every operation is logged |
| **Upserts (MERGE)** | Update + Insert in one operation | Uses log to track row-level changes |

---

### ACID Transactions - What Do They Mean?

| Letter | Word | Simple meaning |
|--------|------|---------------|
| **A** | Atomicity | All of the write succeeds, OR none of it does. No half-writes. |
| **C** | Consistency | Data always moves from one valid state to another valid state |
| **I** | Isolation | Two jobs writing at the same time do not interfere with each other |
| **D** | Durability | Once written and confirmed, data is never lost |

**INSTRUCTOR NOTE:**  
Ask: *'If a job writes 1 million rows and crashes after 500,000 rows, what happens with CSV? What happens with Delta?'*  
CSV: 500,000 rows are written permanently. The data is incomplete.  
Delta: The transaction was not committed. Zero rows are visible. The next read sees clean data.

## Section 3: Delta Table vs CSV/Parquet - Side by Side

**INSTRUCTOR NOTE:**  
This table is a great visual to show on screen. Reference it when students ask 'why not just use Parquet?'

---

| Feature | CSV | Parquet | Delta Table |
|---------|-----|---------|-------------|
| **ACID transactions** | No | No | Yes |
| **Schema enforcement** | No | Partial | Yes |
| **UPDATE rows** | No | No | Yes |
| **DELETE rows** | No | No | Yes |
| **MERGE (upsert)** | No | No | Yes |
| **Time travel** | No | No | Yes |
| **Audit history** | No | No | Yes |
| **Query with SQL** | Limited | Yes | Yes |
| **Handles large data** | Poor | Yes | Yes |
| **Compaction/optimize** | No | Manual | Yes (OPTIMIZE) |
| **Streaming + Batch** | Batch only | Batch only | Both |

**In one sentence:** Delta = Parquet files + a transaction log. That transaction log changes everything.

---

### GlobalMart's Layer-to-Format mapping:

| Layer | Format | Reason |
|-------|--------|--------|
| **Bronze** (`raw/`) | CSV (original files) | Keep raw data exactly as received |
| **Silver** | Delta | Need to clean, update, enforce schema |
| **Gold** | Delta | Need to update aggregations, query fast |

**INSTRUCTOR NOTE:**  
Emphasize: *'From Silver onwards, EVERYTHING is Delta. We never write back to CSV after Bronze.'

## Section 4: Live Demo - Writing and Reading a Delta Table

**INSTRUCTOR NOTE:**  
Now run the code live. Walk through each cell slowly. The key moment is Cell 3 where students see the `_delta_log` folder appear after writing. That folder IS Delta Lake - point to it and explain.

````python
# ============================================================
# CELL 1: Setup - connect to ADLS (same as ILT 2)
#
# INSTRUCTOR NOTE:
#   Every notebook starts with this setup block.
#   Students will eventually put this in a shared config file,
#   but for now we repeat it in each notebook.
# ============================================================

storage_account_name = "amazonprojectadls"
container_name       = "amazon-data"
raw_folder           = "raw"
storage_account_key  = "YOUR_STORAGE_ACCOUNT_KEY"   # paste your key here

# Authenticate Spark to your ADLS account
spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

# Define base paths for each layer
base = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net"
bronze_path = f"{base}/{raw_folder}"      # where raw CSVs live
silver_path = f"{base}/silver"            # where cleaned Delta tables will go
gold_path   = f"{base}/gold"              # where business Delta tables will go

print("Setup complete!")
print(f"  Bronze (raw CSVs) : {bronze_path}")
print(f"  Silver (Delta)    : {silver_path}")
print(f"  Gold   (Delta)    : {gold_path}")
````

````python
# ============================================================
# CELL 2: Read customers from Bronze (raw CSV)
#
# INSTRUCTOR NOTE:
#   This is our Bronze layer - raw CSV exactly as uploaded.
#   Notice: we just read it, we do NOT transform anything here.
#   Bronze is read-only. We only write to Bronze during ingestion.
# ============================================================

# Read the raw customers CSV from bronze
df_bronze_customers = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(f"{bronze_path}/customers.csv")
)

print(f"Bronze customers - Row count : {df_bronze_customers.count():,}")
print(f"Bronze customers - Columns   : {df_bronze_customers.columns}")
print()
print("First 5 rows (raw, no transformation):")
df_bronze_customers.show(5, truncate=True)
````

````python
# ============================================================
# CELL 3: Write customers as a Delta table to Silver
#
# INSTRUCTOR NOTE:
#   This is the key moment - we are CONVERTING a CSV into a Delta table.
#   After running this cell:
#   1. Go to your ADLS container in Azure Portal
#   2. Look inside silver/customers/
#   3. You will see .parquet files AND a _delta_log folder
#   That _delta_log folder = Delta Lake is working!
#
#   format('delta')       = save as Delta format (not CSV or Parquet)
#   mode('overwrite')     = if the table already exists, replace it
#   save(path)            = write to this ADLS path
# ============================================================

silver_customers_path = f"{silver_path}/customers"

# Write the DataFrame as a Delta table
(
    df_bronze_customers
    .write
    .format("delta")          # THIS is what makes it a Delta table
    .mode("overwrite")        # replace existing data if any
    .save(silver_customers_path)
)

print(f"Customers written as Delta table to: {silver_customers_path}")
print()
print("Check your ADLS container in Azure Portal - you will see:")
print("  silver/customers/_delta_log/    <-- transaction log")
print("  silver/customers/part-*.parquet <-- actual data")
````

````python
# ============================================================
# CELL 4: See what Delta actually created on disk
#
# INSTRUCTOR NOTE:
#   Use dbutils.fs.ls() to show students what files were created.
#   The _delta_log folder is what makes this different from plain Parquet.
#   Show them the JSON log file inside - it records exactly what was written.
# ============================================================

print("Files created inside silver/customers/:")
print()

files = dbutils.fs.ls(silver_customers_path)
for f in files:
    print(f"  {f.name}")

print()
print("The _delta_log/ folder is Delta Lake's transaction log.")
print("Every write, update, delete is recorded there as a JSON file.")
print()

# Also show what's inside the _delta_log
print("Contents of _delta_log/:")
log_files = dbutils.fs.ls(f"{silver_customers_path}/_delta_log")
for f in log_files:
    print(f"  {f.name}  ({f.size} bytes)")
````

````python
# ============================================================
# CELL 5: Read back the Delta table
#
# INSTRUCTOR NOTE:
#   Reading a Delta table is the same as reading any format - just
#   change .csv() to .format('delta').load(path)
#
#   The result is the same as before, but now the data has ACID protection,
#   schema enforcement, and time travel capability.
# ============================================================

# Read the Delta table (not CSV anymore - this is Delta format)
df_silver_customers = (
    spark.read
    .format("delta")             # read as Delta (not csv)
    .load(silver_customers_path)
)

print(f"Silver customers (Delta) - Row count : {df_silver_customers.count():,}")
print()
df_silver_customers.show(5, truncate=True)
print()
df_silver_customers.printSchema()
````

````python
# ============================================================
# CELL 6: View Delta table history (audit log)
#
# INSTRUCTOR NOTE:
#   DESCRIBE HISTORY shows every operation ever done on this Delta table.
#   Each row = one version.
#
#   Right now you will see version 0 (the first write we just did).
#   As we do more writes/updates, more versions will appear here.
#   This is the foundation of TIME TRAVEL - we can go back to any version.
# ============================================================

from delta.tables import DeltaTable

# Get the Delta table object
delta_table = DeltaTable.forPath(spark, silver_customers_path)

# Show the full history
print("Delta Table History (every version ever written):")
print()
delta_table.history().select(
    "version", "timestamp", "operation", "operationParameters"
).show(truncate=False)
````

````python
# ============================================================
# CELL 7: Time Travel - Read a specific version of the Delta table
#
# INSTRUCTOR NOTE:
#   Time travel = reading data as it was at a specific point in time.
#   This works because Delta never deletes old data when you write -
#   it just marks new files as current and keeps old ones.
#
#   Two ways to time travel:
#   1. By version number : .option('versionAsOf', 0)
#   2. By timestamp      : .option('timestampAsOf', '2026-06-01')
#
#   Real-world use: 'Show me the data as it was before yesterday's bad pipeline run'
# ============================================================

# Read version 0 of the customers table (the very first write)
df_version_0 = (
    spark.read
    .format("delta")
    .option("versionAsOf", 0)    # go back to version 0
    .load(silver_customers_path)
)

print("Customers at version 0 (first write):")
print(f"Row count: {df_version_0.count():,}")
df_version_0.show(3)

print()
print("This is Time Travel - reading historical data from Delta.")
print("In a real pipeline: if a bad transformation ran, you time-travel back to the good version.")
````

````python
# ============================================================
# CELL 8: Write all 10 GlobalMart tables to Silver as Delta
#
# INSTRUCTOR NOTE:
#   This is the hands-on prep cell. Run this to write all tables.
#   Students will do this themselves in the hands-on.
#   For now, run it together so they can see the pattern.
#
#   Each table goes from:  raw/<table>.csv  (Bronze)
#                    to:   silver/<table>/   (Silver, Delta format)
# ============================================================

globalmart_tables = [
    "customers",
    "orders",
    "orders_items",
    "products",
    "payments",
    "addresses",
    "returns",
    "suppliers",
    "payment_methods",
    "shipping_tier"
]

print(f"{'Table':<25} {'Rows':>10}  Status")
print("-" * 50)

for table_name in globalmart_tables:
    try:
        # Read from Bronze (raw CSV)
        df = (
            spark.read
            .option("header", "true")
            .option("inferSchema", "true")
            .csv(f"{bronze_path}/{table_name}.csv")
        )
        row_count = df.count()

        # Write to Silver as Delta
        (
            df.write
            .format("delta")
            .mode("overwrite")
            .save(f"{silver_path}/{table_name}")
        )

        print(f"{table_name:<25} {row_count:>10,}  Written to silver as Delta")

    except Exception as e:
        print(f"{table_name:<25} {'':>10}  ERROR: {e}")

print("-" * 50)
print("Done! All tables are now in Silver as Delta tables.")
````
