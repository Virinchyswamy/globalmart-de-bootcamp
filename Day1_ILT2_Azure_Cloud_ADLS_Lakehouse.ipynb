## Section 2: Azure Data Lake Storage Gen2 (ADLS Gen2)

**INSTRUCTOR NOTE:**  
This is the most important storage service for us. Spend 10-12 minutes here. Draw the hierarchy on the board/screen while explaining.

---

**ADLS Gen2** = Azure Data Lake Storage Generation 2

It is a cloud storage service that can store **any type of data** (CSV, JSON, Parquet, images, log files) at **massive scale** and at **very low cost**.  
It is what we use as the foundation of the GlobalMart Lakehouse.

---

### ADLS Gen2 Storage Hierarchy:

Think of it like folders on your computer, but in the cloud:

```
Azure Subscription  (your Azure account)
    |
    +-- Resource Group  (a project folder - e.g., 'globalmart-rg')
            |
            +-- Storage Account  (the main unit - e.g., 'globalmartdatalake')
                    |
                    +-- Container  (top-level folder - like a drive letter)
                    |       +-- bronze
                    |       +-- silver
                    |       +-- gold
                    |
                    +-- Inside each container: folders and files
                            bronze/
                                customers/
                                    customers.csv
                                orders/
                                    orders.csv
                                products/
                                    products.csv
```

---

### What makes ADLS Gen2 special vs regular cloud storage?

| Feature | What it means |
|---------|---------------|
| **Hierarchical Namespace** | Real folder structure (not just fake prefixes like Blob Storage) |
| **High Performance** | Optimized for big data analytics - reads fast even on huge files |
| **Folder-level Security** | Set different access permissions per folder (RBAC + ACLs) |
| **Cheap Storage** | Much cheaper than SQL databases or Data Warehouses |
| **Protocol: abfss://** | Works natively with Spark/Databricks, Hadoop, ADF |

---

### ADLS Gen2 vs Blob Storage:

| Feature | Blob Storage | ADLS Gen2 |
|---------|-------------|----------|
| True folder hierarchy | No | Yes |
| Analytics performance | Slow | Fast |
| File/folder-level security | No | Yes |
| Big data (Spark) compatible | Partial | Fully optimized |
| How to enable | Default | Enable 'Hierarchical Namespace' at creation |

**INSTRUCTOR NOTE:**  
ADLS Gen2 is actually built ON TOP of Blob Storage. The secret is: when you create a Storage Account in Azure and enable **'Hierarchical Namespace'**, it becomes ADLS Gen2. This is a checkbox you will see in the hands-on.

## Section 3: Data Warehouse vs Data Lake vs Lakehouse vs Hybrid

**INSTRUCTOR NOTE:**  
This is the conceptual core of this session. Use the restaurant analogy below. Draw the diagrams. Take 15 minutes here - these concepts will come up repeatedly throughout the bootcamp.

---

### The Restaurant Analogy:

| Architecture | Restaurant Analogy | What it really means |
|-------------|-------------------|---------------------|
| **Data Warehouse** | Fine dining - only fully cooked, plated food served | Only clean, structured, processed data |
| **Data Lake** | Grocery store - raw ingredients everywhere, find what you need | Raw data in any format, hard to query |
| **Lakehouse** | Meal kit service - raw ingredients + cooking instructions included | Cheap storage + SQL capability + ACID transactions |
| **Hybrid** | Restaurant that also has a grocery section attached | Both a DW and a DL, used together |

---

### 1. Data Warehouse

```
Source Systems
     |
     v
ETL Process (clean, transform, restructure)
     |
     v
Data Warehouse (structured SQL tables only)
     |
     v
BI Reports / SQL Queries / Dashboards
```

| What | Detail |
|------|--------|
| **Stores** | Only structured (tabular) data |
| **Format** | SQL tables with fixed schema |
| **Cost** | Expensive - compute and storage are bundled |
| **Query Speed** | Very fast for SQL |
| **Examples** | Azure Synapse Analytics, Snowflake, Amazon Redshift |
| **Problem** | Cannot store JSON, images, or unstructured data. Schema must be designed upfront. Very expensive to scale. |

---

### 2. Data Lake

```
Source Systems
     |
     v
ADLS Gen2 (raw files: CSV, JSON, Parquet, images, logs - stored AS IS)
     |
     v
Spark / Databricks (reads and processes files on demand)
```

| What | Detail |
|------|--------|
| **Stores** | Any data - structured, semi-structured, unstructured |
| **Format** | CSV, JSON, Parquet, images, video, logs - anything |
| **Cost** | Very cheap (object/file storage pricing) |
| **Query Speed** | Slow - no indexing, no transactions |
| **Problem** | No ACID transactions (no guaranteed consistency). Hard to update/delete rows. No schema enforcement. Data quality issues. This is called the 'Data Swamp' problem. |

---

### 3. Data Lakehouse (What We Build for GlobalMart)

```
Source Systems
     |
     v
ADLS Gen2 (cheap storage - stores files)
     |
     v
Delta Lake (a layer ON TOP of files - adds ACID, schema, versioning)
     |
     v
Databricks SQL / Spark / BI Tools (query like a warehouse, store like a lake)
```

| What | Detail |
|------|--------|
| **Stores** | Any data format, stored as files in ADLS |
| **Added Layer** | Delta Lake protocol gives ACID transactions + schema + time travel |
| **Cost** | Cheap storage (lake pricing) |
| **Query Speed** | Fast - Delta adds indexes and statistics |
| **Best of Both** | Cheap like a lake. Reliable and queryable like a warehouse. |
| **Examples** | Databricks Lakehouse Platform, Delta Lake on ADLS |

---

### 4. Hybrid Architecture

Some companies keep BOTH a Data Warehouse AND a Data Lake running together:

- **Lake side:** Raw data, ML training data, unstructured files, logs
- **Warehouse side:** Business reporting dashboards, Finance, Sales
- **When is it used?** When a company already has a Data Warehouse but needs to expand to handle more data types, ML, or streaming

---

### Side-by-Side Comparison:

| | Data Warehouse | Data Lake | Lakehouse | Hybrid |
|-|----------------|-----------|-----------|--------|
| **Storage Cost** | High | Low | Low | Medium |
| **Data Types** | Structured only | All types | All types | All types |
| **ACID Transactions** | Yes | No | Yes (Delta) | Partial |
| **Schema Enforcement** | Yes (rigid) | No | Yes (flexible) | Partial |
| **ML / AI Ready** | No | Yes | Yes | Yes |
| **Query Performance** | High | Low | High | Medium |
| **Complexity** | Low | Medium | Medium | High |

---

**INSTRUCTOR NOTE:**  
Ask the class: *'If GlobalMart receives data as CSV files, JSON from APIs, and real-time event streams - which architecture would you recommend and why?'*  
Lead them to: **Lakehouse** - because it handles all three formats, keeps cost low, and gives ACID guarantees via Delta Lake.

## Section 5: Connecting Databricks to ADLS Gen2 - Live Demo

**INSTRUCTOR NOTE:**  
Now switch to live code. Run each code cell on your cluster. Explain what each cell does before running it.  
There are 4 methods to connect Databricks to ADLS Gen2. For this bootcamp, we use **Method 1 (Storage Account Key)** - it is the simplest.

| Method | Security | Best for |
|--------|----------|----------|
| **Storage Account Key** | Low (key = full access) | Learning / Dev |
| **SAS Token** | Medium (time-limited) | Sharing access temporarily |
| **Service Principal** | High | Production environments |
| **Unity Catalog** | Highest (admin-managed) | Enterprise / Team setups |

**Important:** In production, you would NEVER use a Storage Account Key in your notebook code. You would use Key Vault or Unity Catalog. But for learning, Method 1 is fine.

````python
# ============================================================
# CELL 1: Verify your cluster is running
#
# INSTRUCTOR NOTE:
#   Always run this cell first before anything else.
#   If you see 'Cluster is ready!' and the Spark version, your cluster is up.
#   If this fails, the cluster is not started - start it from the Databricks sidebar.
# ============================================================

print("Cluster is ready!")
print(f"Spark version : {spark.version}")
print(f"Python        : working")
````

````python
# ============================================================
# CELL 2: Set your ADLS Gen2 connection details
#
# INSTRUCTOR NOTE:
#   Replace the placeholder values below with your actual Azure values.
#
#   WHERE TO FIND THESE IN AZURE PORTAL:
#   storage_account_name : Azure Portal > Storage Accounts > (your account name)
#   storage_account_key  : Azure Portal > Storage Accounts > Security > Access Keys > key1
#
#   Students will get their values from the hands-on (they create the storage account there).
#   For your demo, use YOUR storage account.
# ============================================================

# Replace with your actual Azure Storage Account name
storage_account_name = "amazonprojectadls"    # e.g., 'globalmartdatalake'

# Replace with your actual Storage Account Access Key (from Azure Portal)
storage_account_key  = "YOUR_STORAGE_ACCOUNT_KEY"     # long string of letters and numbers

print(f"Storage Account : {storage_account_name}")
print(f"Key             : {'*' * 20} (hidden for security)")
````

````python
# ============================================================
# CELL 3: Tell Spark how to authenticate with your ADLS account
#
# INSTRUCTOR NOTE:
#   This is the most important step. We are telling Spark:
#   'When you see files from this storage account, use this key to authenticate.'
#
#   'abfss' = Azure Blob File System Secure (this is the protocol for ADLS Gen2)
#   Think of it like https:// but for Azure storage.
#
#   You MUST run this cell before reading or writing any files from ADLS.
#   If you restart your cluster, you must run this cell again.
# ============================================================

# Register the storage account key with Spark
spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

print("Spark is now configured to connect to ADLS Gen2.")
print(f"Spark will use the key to authenticate to: {storage_account_name}.dfs.core.windows.net")
print("You can now read and write files using the abfss:// path format.")
````

````python
# ============================================================
# CELL 4: Build the ADLS path and verify the connection
#
# INSTRUCTOR NOTE:
#   The ADLS Gen2 URL format is:
#   abfss://<container_name>@<storage_account>.dfs.core.windows.net/<folder_path>
#
#   Example:
#   abfss://bronze@globalmartdatalake.dfs.core.windows.net/customers/customers.csv
#
#   dbutils.fs.ls() = lists files in a folder (like 'ls' command in Linux)
#   If you see your files listed, the connection is working!
# ============================================================

# Build the path to the bronze container
bronze_path = f"abfss://bronze@{storage_account_name}.dfs.core.windows.net/"

print(f"Bronze container path: {bronze_path}")
print()

# List files/folders inside the bronze container
try:
    items = dbutils.fs.ls(bronze_path)
    if items:
        print(f"Connection successful! Found {len(items)} items in bronze container:")
        for item in items:
            print(f"  - {item.name}  (size: {item.size} bytes)")
    else:
        print("Connection successful! The bronze container is empty.")
        print("Upload the GlobalMart CSV files in the hands-on, then rerun this cell.")
except Exception as e:
    print(f"Connection failed!")
    print(f"Error: {e}")
    print()
    print("Check these:")
    print("  1. Is storage_account_name correct?")
    print("  2. Is storage_account_key correct?")
    print("  3. Did you run Cell 3 to configure Spark?")
    print("  4. Does the 'bronze' container exist in your storage account?")
````

````python
# ============================================================
# CELL 5: Read the customers.csv file from ADLS Bronze container
#
# INSTRUCTOR NOTE:
#   This is the key payoff cell - students see data appearing from the cloud!
#
#   spark.read.csv() = reads a CSV file and creates a Spark DataFrame
#   header=True      = the first row of the CSV contains column names
#   inferSchema=True = Spark will automatically detect data types (int, string, date)
#
#   df.show(5)   = shows the first 5 rows in a table format
#   df.count()   = returns the total number of rows
#
#   NOTE: Run this AFTER uploading customers.csv to bronze container in hands-on.
# ============================================================

# Path to the customers file in bronze
customers_path = f"abfss://bronze@{storage_account_name}.dfs.core.windows.net/customers/customers.csv"

# Read the CSV file into a Spark DataFrame
df_customers = (
    spark.read
    .option("header", "true")       # first row = column names
    .option("inferSchema", "true")  # auto-detect data types
    .csv(customers_path)
)

# Show results
print(f"Total customers loaded: {df_customers.count():,}")
print()
df_customers.show(5, truncate=False)
````

````python
# ============================================================
# CELL 6: Explore the schema of the customers table
#
# INSTRUCTOR NOTE:
#   printSchema() = shows column names and their data types
#   This is the FIRST thing you should run on any new dataset.
#   It tells you what columns exist and what type of data they contain.
#
#   Common Spark data types:
#   StringType  = text
#   IntegerType = whole numbers
#   DoubleType  = decimal numbers
#   TimestampType = date + time
#   BooleanType = true/false
# ============================================================

print("=== CUSTOMERS TABLE SCHEMA ===")
df_customers.printSchema()

print(f"\nRows    : {df_customers.count():,}")
print(f"Columns : {len(df_customers.columns)}")
print(f"Names   : {df_customers.columns}")
````

````python
# ============================================================
# CELL 7: Read all 10 GlobalMart tables and show row counts
#
# INSTRUCTOR NOTE:
#   Run this cell AFTER students upload all 10 CSV files in the hands-on.
#   This is a great way to verify everything is uploaded and readable.
#   If a file is missing, it will print 'Not found yet' instead of failing.
# ============================================================

# All 10 GlobalMart source tables
globalmart_tables = [
    "customers",
    "orders",
    "orders_items",
    "products",
    "payments",
    "addresses",
    "returns",
    "suppliers",
    "payment_methods",
    "shipping_tier"
]

base_path = f"abfss://bronze@{storage_account_name}.dfs.core.windows.net"

print(f"{'Table':<25} {'Row Count':>12}")
print("-" * 40)

for table_name in globalmart_tables:
    try:
        file_path = f"{base_path}/{table_name}/{table_name}.csv"
        df = spark.read.option("header", "true").csv(file_path)
        row_count = df.count()
        print(f"{table_name:<25} {row_count:>12,}")
    except Exception:
        print(f"{table_name:<25} {'Not found yet':>12}")

print("-" * 40)
print("Done! All available tables read from ADLS Bronze container.")
````
