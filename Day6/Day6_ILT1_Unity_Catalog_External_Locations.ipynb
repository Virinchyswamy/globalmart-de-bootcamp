## 1. The Problem with Hardcoded Storage Keys

Until now, every notebook that reads from ADLS has started with this pattern:

```python
storage_account_name = "amazonprojectadls"
storage_account_key  = "YOUR_STORAGE_ACCOUNT_KEY"   # ← 88-character secret key

spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)
```

### Why This Is a Problem

| Problem | What It Means |
|---------|---------------|
| **Security risk** | Anyone who opens the notebook sees the full storage key — gives read/write access to ALL data in the account |
| **Git exposure** | If the notebook is committed to GitHub, the key is permanently in the git history even if deleted later |
| **Key rotation** | When the storage key is rotated (required by security policy), every notebook that hardcodes it breaks |
| **No audit trail** | Azure has no way to know *which user* accessed the data — it just sees "key was used" |
| **Shared secret** | All notebooks share the same key — you cannot give one team read-only and another team read-write |

### The Root Cause

```
CURRENT FLOW (key-based):

  Notebook → spark.conf.set(key) → Spark driver → ADLS
                    ↑
              key lives in plain text
              in the notebook cell
              visible to anyone with notebook access

WHAT WE WANT:

  Notebook → abfss://... → Unity Catalog checks permission → ADLS
                                        ↑
                              no key anywhere
                              auth handled centrally
```

---
## 2. The Unity Catalog Solution — Two Building Blocks

Unity Catalog replaces the key with two objects that live in Databricks:

---

### Building Block 1 — Storage Credential

```
A Storage Credential is Databricks' identity card for Azure.

It wraps an Azure Managed Identity (the Databricks Access Connector).
The Access Connector is an Azure resource that has been granted
  Storage Blob Data Contributor on your ADLS storage account.

When Databricks needs to read/write ADLS, it uses the managed identity
to authenticate — no password, no key, no secret to leak.
```

**Analogy:** The Storage Credential is like an employee badge — it says "this is who Databricks is" to Azure.

---

### Building Block 2 — External Location

```
An External Location is a registered ADLS path + which Storage Credential to use.

  External Location: amazon_data_external
    URL:                abfss://amazon-data@amazonprojectadls.dfs.core.windows.net/
    Storage Credential: amazonprojectadls_credential

Unity Catalog uses this mapping every time a notebook accesses that path.
```

**Analogy:** The External Location is the door that the badge unlocks — it maps a specific room (ADLS path) to the badge (credential).

---

### The Full Chain

```
Notebook uses abfss://amazon-data@amazonprojectadls.dfs.core.windows.net/raw/
       ↓
Unity Catalog: "I know this path — it's covered by External Location 'amazon_data_external'"
       ↓
Storage Credential: "Use managed identity 'amazonprojectadls_credential'"
       ↓
Azure RBAC: "Managed identity has Storage Blob Data Contributor on amazonprojectadls"
       ↓
ADLS Gen2: access granted — data returned

At no point does a key appear anywhere.
```

## 3. Prerequisites — RBAC Roles on Azure

Before creating the Storage Credential in Databricks, the Azure side must be configured.
Your manager has already done this — here is what was granted and why each role matters:

| Role | Granted On | Why It's Needed |
|------|-----------|------------------|
| **Storage Blob Data Contributor** | Storage Account (`amazonprojectadls`) | Allows the managed identity to read and write blobs (actual data files) |
| **Storage Account Contributor** | Storage Account (`amazonprojectadls`) | Allows Databricks to manage the storage account settings (needed for some Unity Catalog operations) |
| **EventGrid EventSubscription Contributor** | Resource Group | Allows Databricks to create event subscriptions for Auto Loader file notifications |

```
Azure Portal structure:

  Resource Group (your-rg)
    ├── Storage Account: amazonprojectadls
    │     ├── Role: Storage Blob Data Contributor → Databricks Access Connector
    │     └── Role: Storage Account Contributor   → Databricks Access Connector
    │
    └── Role: EventGrid EventSubscription Contributor → Databricks Access Connector

  Databricks Workspace
    └── Access Connector (managed identity) ← receives all 3 roles above
```

> **Check:** In Azure Portal → Storage Account → Access Control (IAM) → Role assignments → you should see your Access Connector listed under all 3 roles.

## 4. SETUP — Step 1: Find the Access Connector Resource ID

The Access Connector resource ID is what you paste into Databricks when creating the Storage Credential.

### In Azure Portal:

```
1. Azure Portal → top search bar → type: "Access Connectors for Azure Databricks"
2. Click on your Access Connector (linked to your Databricks workspace)
3. Left sidebar → Settings → Properties
4. Copy the Resource ID field — it looks like:

   /subscriptions/xxxxxxxx-xxxx-xxxx-xxxx-xxxxxxxxxxxx
   /resourceGroups/your-resource-group
   /providers/Microsoft.Databricks
   /accessConnectors/your-connector-name

   (one long string — copy all of it)
```

### If You Don't See an Access Connector:

```
Option A — It may be named differently:
  Search for the Access Connector by your Databricks workspace name.
  It is often created automatically when the workspace was provisioned.

Option B — Create one:
  Azure Portal → Create a resource → search "Azure Databricks Access Connector"
  Create it in the same Resource Group as your Databricks workspace
  Then re-apply the RBAC roles to this new connector
```

> **Save the Resource ID** — you will paste it in the next step.

## 9. Granting Access to Other Users

Unity Catalog lets you control who can use the External Location.
By default, only the creator (you) can use it — grant access to teammates via SQL:

```sql
-- Grant read access to a specific user:
GRANT READ FILES ON EXTERNAL LOCATION amazon_data_external
TO `user@yourcompany.com`;

-- Grant read + write access:
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION amazon_data_external
TO `user@yourcompany.com`;

-- Grant to a group:
GRANT READ FILES ON EXTERNAL LOCATION amazon_data_external
TO `data-engineers`;

-- See current grants:
SHOW GRANTS ON EXTERNAL LOCATION amazon_data_external;
```

### Why This Matters vs the Key Approach

```
KEY APPROACH:
  Anyone with the key has full read/write access to the entire storage account.
  You cannot give read-only access. You cannot revoke access without rotating the key.
  Rotation breaks every notebook that hardcodes the old key.

UNITY CATALOG APPROACH:
  Grant read-only to analysts.
  Grant read+write to engineers.
  Revoke access instantly without touching ADLS or rotating anything.
  Full audit log: Unity Catalog records every access, by which user, at what time.
```

## 11. Summary — What Unity Catalog Gives Us

```
WITHOUT UNITY CATALOG (key-based):

  Security:   Key in notebook → leaked if notebook is shared or committed to git
  Rotation:   Key rotated → all notebooks break simultaneously
  Audit:      No record of who accessed what — just "key was used"
  Governance: Cannot restrict access to specific paths or users

WITH UNITY CATALOG (external location):

  Security:   No key in notebook — managed identity handles auth invisibly
  Rotation:   Nothing to rotate — managed identity tokens are short-lived and automatic
  Audit:      Full audit log in Unity Catalog — user, path, timestamp, query
  Governance: Grant READ/WRITE per user, per group, per path
```

---

## Key Takeaways

1. **Storage Credential** = Databricks' registered managed identity — the "who" that talks to Azure
2. **External Location** = a registered ADLS path + which credential to use — the "what" that is accessible
3. Once set up, `abfss://` paths work in any notebook — **no `spark.conf.set`, no key, no setup cell**
4. Access is controlled centrally — grant per user, revoke instantly, full audit trail
5. Only ADLS is covered by this — Supabase / Neo4j credentials are handled separately (Databricks Secrets)
