# Task 2 — ADLS → Autoloader → Bronze Delta (Customers + Payments)

**Goal:** Use Databricks Autoloader to read `customers.csv` and `payments.csv` from ADLS
and land them in Bronze as Delta tables.

**What makes Autoloader special:**
- It **automatically detects new files** dropped in the ADLS folder
- It **remembers which files it already processed** (no duplicates)
- It **infers schema** automatically — you don't have to define columns
- It is built for **incremental ingestion** — production-grade from day one

---
## What is Autoloader?

```
Without Autoloader (manual approach):
  You read all files every time → expensive
  Or you manually track which files are new → complex code

With Autoloader:
  Run 1: ADLS has customers_jan.csv  → Autoloader reads it, saves checkpoint
  Run 2: ADLS has customers_feb.csv  → Autoloader reads ONLY the new file
  Run 3: No new files                → Autoloader does nothing (very fast)
```

**Autoloader = incremental file ingestion made easy.**

---
## Key Autoloader Concepts

| Term | What it means |
|------|---------------|
| `cloudFiles` | The Autoloader format — like saying `format("delta")` but for reading files incrementally |
| `cloudFiles.format` | What kind of files you're reading — `csv`, `json`, `parquet` |
| `cloudFiles.schemaLocation` | Where Autoloader stores the inferred schema — so it's consistent across runs |
| `checkpointLocation` | Where Autoloader tracks which files have been processed — the "memory" |
| `trigger(availableNow=True)` | Process all available new files and stop — like a batch job |
| `trigger(processingTime='1 minute')` | Process new files every 1 minute — like a streaming job |

````python
# ─── Setup: ADLS connection ───────────────────────────────────────────────────

storage_account_name = "amazonprojectadls"
container_name       = "amazon-data"
storage_account_key  = "YOUR_STORAGE_ACCOUNT_KEY"  # ← Replace

spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

raw_path    = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/raw"
bronze_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/bronze"

# Autoloader needs two special folders:
# 1. schemaLocation — stores the schema it inferred (once per table)
# 2. checkpointLocation — tracks which files were already processed
autoloader_base = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/_autoloader"

print("ADLS connected!")
print(f"Source (raw) : {raw_path}")
print(f"Target (bronze): {bronze_path}")
print(f"Autoloader metadata: {autoloader_base}")
````

---
## Part 1 — Ingest customers.csv with Autoloader

````python
# ─── Step 1: Read customers.csv using Autoloader (cloudFiles) ─────────────────
# This does NOT read the file yet — it sets up the stream definition

customers_schema_path     = f"{autoloader_base}/schemas/customers"
customers_checkpoint_path = f"{autoloader_base}/checkpoints/customers"

customers_stream = spark.readStream \
    .format("cloudFiles") \
    .option("cloudFiles.format",         "csv") \
    .option("cloudFiles.schemaLocation", customers_schema_path) \
    .option("header",                    "true") \
    .option("inferSchema",               "true") \
    .load(f"{raw_path}/customers.csv")

# Add audit columns — track when and where the file came from
from pyspark.sql.functions import current_timestamp, input_file_name, lit

customers_stream = customers_stream \
    .withColumn("_ingested_at",  current_timestamp()) \
    .withColumn("_source_file",  input_file_name()) \
    .withColumn("_source",       lit("adls_autoloader"))

print("Autoloader stream defined for customers.")
print("Schema will be inferred from the file automatically.")
````

````python
# ─── Step 2: Write customers to Bronze as Delta ───────────────────────────────
# trigger(availableNow=True) = process all current files and stop
# This is the batch mode — like a scheduled job that runs once

customers_bronze_path = f"{bronze_path}/customers"

customers_query = customers_stream.writeStream \
    .format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", customers_checkpoint_path) \
    .trigger(availableNow=True) \
    .start(customers_bronze_path)

# Wait for it to finish
customers_query.awaitTermination()

# Verify
customers_count = spark.read.format("delta").load(customers_bronze_path).count()
print(f"Customers in Bronze: {customers_count:,} rows")
print(f"Location: {customers_bronze_path}")
````

````python
# ─── Step 3: Inspect the result ──────────────────────────────────────────────

customers_bronze = spark.read.format("delta").load(customers_bronze_path)

print("Schema inferred by Autoloader:")
customers_bronze.printSchema()

print("\nSample rows (with audit columns):")
customers_bronze.select("*").show(5, truncate=True)

print("\nAudit columns added:")
customers_bronze.select("_ingested_at", "_source_file", "_source").show(3, truncate=False)
````

---
## Part 2 — Ingest payments.csv with Autoloader

````python
# ─── Read payments.csv using Autoloader ──────────────────────────────────────

payments_schema_path     = f"{autoloader_base}/schemas/payments"
payments_checkpoint_path = f"{autoloader_base}/checkpoints/payments"
payments_bronze_path     = f"{bronze_path}/payments"

payments_stream = spark.readStream \
    .format("cloudFiles") \
    .option("cloudFiles.format",         "csv") \
    .option("cloudFiles.schemaLocation", payments_schema_path) \
    .option("header",                    "true") \
    .option("inferSchema",               "true") \
    .load(f"{raw_path}/payments.csv")

payments_stream = payments_stream \
    .withColumn("_ingested_at", current_timestamp()) \
    .withColumn("_source_file", input_file_name()) \
    .withColumn("_source",      lit("adls_autoloader"))

# Write to Bronze
payments_query = payments_stream.writeStream \
    .format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", payments_checkpoint_path) \
    .trigger(availableNow=True) \
    .start(payments_bronze_path)

payments_query.awaitTermination()

payments_count = spark.read.format("delta").load(payments_bronze_path).count()
print(f"Payments in Bronze: {payments_count:,} rows")
print(f"Location: {payments_bronze_path}")
````

````python
# ─── Inspect payments ─────────────────────────────────────────────────────────

payments_bronze = spark.read.format("delta").load(payments_bronze_path)

print("Payments schema:")
payments_bronze.printSchema()

print("\nSample rows:")
payments_bronze.show(5, truncate=True)
````

---
## Part 3 — The MAGIC of Autoloader: Run it again and see what happens

````python
# ─── Run Autoloader again on the SAME file ────────────────────────────────────
# Expectation: Autoloader should detect that customers.csv was ALREADY processed
# and skip it — no duplicate rows in Bronze

print("Running Autoloader on customers.csv again...")

customers_before = spark.read.format("delta").load(customers_bronze_path).count()
print(f"Customers in Bronze BEFORE second run: {customers_before:,}")

# Re-run the same Autoloader stream
customers_stream2 = spark.readStream \
    .format("cloudFiles") \
    .option("cloudFiles.format",         "csv") \
    .option("cloudFiles.schemaLocation", customers_schema_path) \
    .option("header",                    "true") \
    .option("inferSchema",               "true") \
    .load(f"{raw_path}/customers.csv")

customers_stream2 = customers_stream2 \
    .withColumn("_ingested_at", current_timestamp()) \
    .withColumn("_source_file", input_file_name()) \
    .withColumn("_source",      lit("adls_autoloader"))

customers_query2 = customers_stream2.writeStream \
    .format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", customers_checkpoint_path) \
    .trigger(availableNow=True) \
    .start(customers_bronze_path)

customers_query2.awaitTermination()

customers_after = spark.read.format("delta").load(customers_bronze_path).count()
print(f"Customers in Bronze AFTER second run : {customers_after:,}")
print()

if customers_after == customers_before:
    print("File was already processed — Autoloader SKIPPED it!")
    print("No duplicates added.")
else:
    print(f"New rows added: {customers_after - customers_before:,}")
````

````python
# ─── Check the checkpoint to understand HOW Autoloader tracks files ───────────
# The checkpoint folder contains metadata about every file that was processed
# Autoloader checks this before processing — if a file is listed, it skips it

print("Contents of the Autoloader checkpoint folder:")
dbutils.fs.ls(customers_checkpoint_path)

# You'll see folders like:
#   commits/         ← list of completed micro-batches
#   offsets/         ← tracks progress through the source
#   sources/         ← tracks which exact files have been read
#   metadata         ← query metadata

# This is why Autoloader never re-reads the same file:
# Every file's path and size is recorded in the checkpoint after processing.
````

## Part 4 — Bonus: Read an entire folder (all CSVs at once)

````python
# ─── Read ALL CSV files in raw/ with one Autoloader stream ───────────────────
# Instead of pointing to a single file, point to the whole raw/ folder
# Autoloader will process ALL .csv files in that folder
# When a new CSV is dropped later, Autoloader picks it up on the next run

# This is the PRODUCTION pattern — one Autoloader per table (or per folder)

all_files_schema_path     = f"{autoloader_base}/schemas/all_raw"
all_files_checkpoint_path = f"{autoloader_base}/checkpoints/all_raw"
all_files_bronze_path     = f"{bronze_path}/all_raw_combined"

all_files_stream = spark.readStream \
    .format("cloudFiles") \
    .option("cloudFiles.format",             "csv") \
    .option("cloudFiles.schemaLocation",     all_files_schema_path) \
    .option("header",                        "true") \
    .option("inferSchema",                   "true") \
    .option("cloudFiles.schemaEvolutionMode","addNewColumns") \
    .load(raw_path)  # <-- whole folder, not a single file

all_files_stream = all_files_stream \
    .withColumn("_ingested_at", current_timestamp()) \
    .withColumn("_source_file", input_file_name())

# NOTE: Combining different CSVs into one table only works if they have similar schema
# For production: run separate Autoloader per table (one per CSV)
# This demo just shows that folder-level reading is possible

all_files_query = all_files_stream.writeStream \
    .format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", all_files_checkpoint_path) \
    .trigger(availableNow=True) \
    .start(all_files_bronze_path)

all_files_query.awaitTermination()

total = spark.read.format("delta").load(all_files_bronze_path).count()
print(f"Total rows from all CSVs combined: {total:,}")
````

---
## Summary — Autoloader vs Manual CSV Read

| | Manual `spark.read.csv` | Autoloader (`cloudFiles`) |
|-|------------------------|---------------------------|
| **Tracks processed files** | No — reads everything every time | Yes — checkpoint remembers every file |
| **Handles new files** | You must add code to detect them | Automatic — detects new files on next run |
| **Schema inference** | Once per run | Inferred once, stored, reused across runs |
| **Schema evolution** | Breaks if a new column appears | `addNewColumns` mode handles it gracefully |
| **Streaming support** | No | Yes — can run as a continuous stream |
| **Duplicates risk** | High (re-reads on every run) | Zero (checkpoint prevents re-reads) |
| **When to use** | Quick one-off read | Production ingestion pipeline |
