## Section 1: Why Do Data Engineers Need Version Control?

**INSTRUCTOR NOTE:**  
Start with a relatable story. Use the scenario below.

---

### The Problem Without Version Control:

Imagine this scenario at GlobalMart:

- Monday: You write a notebook that cleans customer data and it works perfectly
- Tuesday: You make changes to add a new column
- Wednesday: The pipeline breaks. The new column had a bug.
- **What do you do?** You cannot go back to Monday's working version because you overwrote it.

Or this scenario:
- Two engineers are both editing the same notebook at the same time
- One saves their version, the other saves their version — one person's work is gone

**Version control solves both problems.**

---

### What Version Control Does for You:

| Problem | How Version Control Fixes It |
|---------|-----------------------------|
| Accidentally broke working code | Roll back to any previous version instantly |
| Two engineers editing same file | Merge changes safely, see who changed what |
| Need to audit what changed | Full history of every change ever made |
| Deploying to production safely | Test in dev/test branch before touching prod |
| Onboarding a new team member | They clone the repo and have everything immediately |

**INSTRUCTOR NOTE:**  
Ask students: *'Has anyone ever accidentally deleted or overwritten code and lost work? That never happens with Git.'

## Section 2: Git Basics (Very Simple)

**INSTRUCTOR NOTE:**  
Keep this very simple. Students only need to understand 6 concepts. Draw the diagram on the board.

---

**Git** is the most popular version control tool in the world.  
**GitHub** is a website where you store your Git repositories online (like Google Drive but for code).

---

### 6 Git Concepts You Need to Know:

| Concept | Simple Meaning | Real Life Analogy |
|---------|---------------|-------------------|
| **Repository (Repo)** | A folder that tracks all changes to all files inside it | A project folder with a complete history |
| **Branch** | A separate copy of the code you can work on safely | A rough draft that does not affect the published version |
| **Commit** | Save a snapshot of your current changes with a message | Pressing Save + writing a note about what you saved |
| **Push** | Upload your commits from Databricks to GitHub | Syncing your phone photos to Google Photos |
| **Pull** | Download latest changes from GitHub into Databricks | Refreshing your email to get new messages |
| **Merge** | Combine changes from one branch into another | Combining two documents into one final version |

---

### How It All Flows:

```
GitHub (online)
    |
    |  <-- push (upload your changes)
    |  --> pull (download latest changes)
    |
Databricks Repos (your local working copy)
    |
    +-- dev branch     (where you write new code)
    +-- test branch    (where you test before going live)
    +-- prod branch    (the live, running pipeline - only tested code here)
```

**INSTRUCTOR NOTE:**  
Draw this on the board. Tell students: *'GitHub is the source of truth. Databricks Repos is your working copy. Always push your work to GitHub so it is safe and shared.'

## Section 3: The Dev → Test → Prod Branch Strategy

**INSTRUCTOR NOTE:**  
This is the most important practical concept of this session. Every professional data team uses this pattern.

---

In a professional data engineering team, you NEVER write code directly on the production branch.  
You always work through 3 environments:

```
DEV branch                TEST branch              PROD branch
==========                ===========              ===========
Where you                 Where you                The live pipeline.
write and                 test with                Real data.
experiment.               real data                Business depends
Mistakes here             before going             on this.
are okay.                 live.                    Never break this.
     |
     |--- Pull Request --> TEST
                               |
                               |--- Pull Request --> PROD
```

### What Happens in Each Branch:

| Branch | Who works here | What runs here | Data used |
|--------|---------------|---------------|----------|
| **dev** | Individual engineers | New features, bug fixes, experiments | Sample/test data |
| **test** | QA / team review | Integration testing, data quality checks | Full data, read-only |
| **prod** | Automated pipelines only | The final, approved, live pipeline | Live production data |

### The Rule:

> Code only moves forward: **dev → test → prod**  
> It never goes backwards. You never edit prod directly.

### For GlobalMart:

```
globalmart-lakehouse/          <- your GitHub repo
    |
    +-- branch: dev            <- you write here every day
    |       notebooks/
    |           bronze_ingestion.ipynb
    |           silver_customers.ipynb
    |
    +-- branch: test           <- tested and reviewed code
    |
    +-- branch: prod           <- only fully approved code
```

**INSTRUCTOR NOTE:**  
Ask: *'Why would a company care about keeping prod separate from dev?'*  
Answer: If a bug reaches prod, the business dashboards break, Finance gets wrong numbers, reports fail. The cost is very high.

## Section 4: Databricks Repos — Connecting GitHub to Databricks

**INSTRUCTOR NOTE:**  
Now walk students through the actual Databricks UI. Follow the steps below on your screen while explaining.

---

**Databricks Repos** is a built-in feature of Databricks that lets you connect your notebooks directly to a GitHub repository.  
Instead of saving notebooks only inside Databricks, they also live in GitHub — version-controlled and shareable.

### How to Connect GitHub to Databricks — Step by Step:

**Part A: GitHub Setup (do this first)**

1. Go to **github.com** and log in (create a free account if you do not have one)
2. Click **+ New repository**
3. Name it: `globalmart-lakehouse`
4. Set it to **Private**
5. Check **Add a README file**
6. Click **Create repository**
7. Go to **Settings → Developer Settings → Personal Access Tokens → Tokens (classic)**
8. Click **Generate new token (classic)**
9. Give it a name, set expiry to 90 days, check the **repo** scope
10. Click **Generate token** → **COPY the token immediately** (you cannot see it again)

---

**Part B: Databricks Setup**

1. In Databricks, click your username (top right) → **Settings**
2. Go to **Linked Accounts** → **Git Integration**
3. Select **GitHub** as the provider
4. Paste your **Personal Access Token** from GitHub
5. Enter your **GitHub username**
6. Click **Save**

---

**Part C: Clone the Repo into Databricks**

1. In the left sidebar, click **Repos**
2. Click **Add Repo**
3. Paste your GitHub repo URL: `https://github.com/YOUR_USERNAME/globalmart-lakehouse`
4. Click **Create Repo**
5. Your repo now appears in Databricks under Repos
6. Any notebook you create inside this folder is automatically tracked by Git

## Section 5: The Daily Git Workflow in Databricks

**INSTRUCTOR NOTE:**  
This is the workflow students will follow every single day of the bootcamp from today onwards.

---

### Every day when you start working:

```
Step 1: Pull latest changes
   In Databricks Repos → click your repo → click Git button → Pull
   (Get any changes your teammates made)

Step 2: Make sure you are on the dev branch
   In the Git panel → check branch name → switch to dev if needed

Step 3: Write your code in the notebook

Step 4: Commit your changes
   In Databricks Repos → click Git button → Stage changes → Write a message → Commit
   Example commit message: 'Add silver_customers transformation notebook'

Step 5: Push to GitHub
   In the Git panel → Push
   Now your code is safe in GitHub.
```

### Good Commit Message Examples:

| Bad message | Good message |
|------------|-------------|
| `update` | `Add null check for customer email in silver layer` |
| `fix` | `Fix date format in orders table from MM/DD to YYYY-MM-DD` |
| `changes` | `Add bronze ingestion notebook for customers table` |
| `done` | `Complete silver transformation for all 10 GlobalMart tables` |

**Rule:** Your commit message should tell a new team member exactly what changed and why.

## Section 6: Live Demo — Create a Branch and Commit a Notebook

**INSTRUCTOR NOTE:**  
Walk through this live. Students follow on their own Databricks.  
The goal: by end of demo, everyone has created a dev branch and pushed one notebook to GitHub.

````python
# ============================================================
# CELL 1: Verify you are working inside a Git repo
#
# INSTRUCTOR NOTE:
#   Run this to confirm the notebook is inside a Databricks Repo
#   (not just a regular Databricks workspace folder).
#
#   dbutils.notebook.entry_point.getDbutils().notebook().getContext()
#   gives us information about the current notebook.
#
#   If the notebook path contains '/Repos/' it is Git-tracked.
#   If it contains '/Users/' it is NOT in a repo.
# ============================================================

# Get the current notebook's file path
notebook_path = dbutils.notebook.entry_point.getDbutils().notebook().getContext().notebookPath().get()

print(f"Current notebook path: {notebook_path}")
print()

# Check if we are inside a Repo (Git-tracked) or a regular workspace
if "/Repos/" in notebook_path:
    print("This notebook IS inside a Databricks Repo - Git tracking is active.")
    print("Any changes you commit here will be tracked in GitHub.")
else:
    print("This notebook is NOT inside a Repo.")
    print("Move it to your Repos folder to enable Git tracking.")
    print("Repos folder path: /Repos/YOUR_EMAIL/globalmart-lakehouse/")
````

````python
# ============================================================
# CELL 2: Show the recommended folder structure for GlobalMart repo
#
# INSTRUCTOR NOTE:
#   This is just a print cell to show students what the repo
#   folder structure should look like.
#   They will create these folders manually in Databricks Repos.
#   Good folder structure = easy to navigate for the whole team.
# ============================================================

repo_structure = """
globalmart-lakehouse/              <- root of the GitHub repo
    |
    +-- README.md                  <- what this project is about
    |
    +-- notebooks/
    |       |
    |       +-- bronze/            <- raw ingestion notebooks
    |       |       ingest_customers.ipynb
    |       |       ingest_orders.ipynb
    |       |
    |       +-- silver/            <- cleaning and transformation notebooks
    |       |       transform_customers.ipynb
    |       |       transform_orders.ipynb
    |       |
    |       +-- gold/              <- business layer notebooks
    |               monthly_revenue.ipynb
    |               customer_segments.ipynb
    |
    +-- config/                    <- shared config (storage paths, settings)
            setup.ipynb
"""

print("Recommended GlobalMart Repo Structure:")
print(repo_structure)
````

````python
# ============================================================
# CELL 3: Shared config - connection details used across all notebooks
#
# INSTRUCTOR NOTE:
#   Instead of repeating the ADLS connection code in every notebook,
#   we put it in one shared config notebook and run it with %run.
#
#   This is a best practice:
#   - Change the storage key in ONE place -> all notebooks pick it up
#   - Consistent variable names across all notebooks
#   - Clean, professional code
#
#   In the hands-on, students create this config/setup.ipynb file
#   inside their Databricks Repo.
# ============================================================

# This is what config/setup.ipynb will contain:
# (In the real setup, this lives in a separate notebook called by %run)

# ---- GLOBALMART SHARED CONFIG ----

# Azure ADLS connection details
STORAGE_ACCOUNT_NAME = "amazonprojectadls"
CONTAINER_NAME       = "amazon-data"
STORAGE_KEY          = "YOUR_STORAGE_ACCOUNT_KEY"   # paste your key here

# Authenticate Spark to ADLS
spark.conf.set(
    f"fs.azure.account.key.{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net",
    STORAGE_KEY
)

# Define standard paths for each layer
BASE_PATH    = f"abfss://{CONTAINER_NAME}@{STORAGE_ACCOUNT_NAME}.dfs.core.windows.net"
BRONZE_PATH  = f"{BASE_PATH}/raw"       # raw CSV files land here
SILVER_PATH  = f"{BASE_PATH}/silver"    # cleaned Delta tables go here
GOLD_PATH    = f"{BASE_PATH}/gold"      # business-ready Delta tables go here

print("GlobalMart config loaded successfully.")
print(f"  Bronze path : {BRONZE_PATH}")
print(f"  Silver path : {SILVER_PATH}")
print(f"  Gold path   : {GOLD_PATH}")
````

````python
# ============================================================
# CELL 4: How to USE the shared config in any other notebook
#
# INSTRUCTOR NOTE:
#   Show students that instead of copy-pasting the config
#   into every notebook, they use %run to call it.
#
#   %run = runs another notebook and imports all its variables
#   into the current notebook.
#
#   After %run, BRONZE_PATH, SILVER_PATH, GOLD_PATH are all
#   available in this notebook without re-defining them.
# ============================================================

# In any notebook, instead of repeating all the config code, just write:
# %run ../config/setup

# That single line loads all the shared config variables.
# The path '../config/setup' means: go up one folder, then into config/setup notebook.

# For demonstration, we define them here manually:
print("In real notebooks, we will start with:")
print()
print('%run ../config/setup')
print()
print("That one line replaces all the ADLS connection code.")
print("Then BRONZE_PATH, SILVER_PATH, GOLD_PATH are ready to use immediately.")
````
